2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]

```python
# calculando as estatísticas e criando uma pivot table
df_saude_sono.pivot_table(values="Duração do sono",index="Profissão", aggfunc=[np.mean,np.median,pd.Series.mode])
```

**ou usando o groupby:**

```python
# calculando as estatísticas apenas com agrupamento(no caso da moda, pela possibilidade de exibir mais de um resultado, o tolist() serve para criar uma lista com os resultado)
resultado = df_saude_sono.groupby("Profissão")["Duração do sono"].agg(mean="mean",median=np.median,mode=lambda x: pd.Series.mode(x).tolist())
resultado
```

4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)



```python
# calculando a média da duração de sono para advogados
media_sono_advogado=df_saude_sono[df_saude_sono["Profissão"].isin(["Advogado(a)"])]["Duração do sono"].mean()
media_sono_advogado
```

```python
# calculando a média da duração de sono para os representantes de vendas
media_sono_representante=df_saude_sono[df_saude_sono["Profissão"].isin(["Representante de Vendas"])]["Duração do sono"].mean()
media_sono_representante
```

```python
# calculando a média de sono para os representantes de outras profissões
media_sono_outras_profissoes=df_saude_sono[~df_saude_sono["Profissão"].isin(["Advogado(a)","Representante de Vendas"])]["Duração do sono"].mean()
media_sono_outras_profissoes
```

Comparando a média de horas dormidas, somente ser Representante de Vendas faz você dormir menos do que dormem pessoas que exercem outras profissões.

***uma outra análise é possível, considerando advogados e representantes como um grupo, e não separadamente:***

```python
# calculando a média de duração do sono para advogados e representantes de vendas considerando-os como um grupo
media_sono_grupo=df_saude_sono[df_saude_sono["Profissão"].isin(["Advogado(a)","Representante de Vendas"])]["Duração do sono"].mean()
media_sono_grupo
```

Considerando desta forma, a média de sono dos advogados e representantes é bem próxima das outras profissões.

8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
# Separando os valores da pressão sistólica e diastólica na coluna de pressão sanguínea, usando '/' como separador
sistolica_diastolica=df_saude_sono["Pressão Sanguínea"].str.split(pat="/",n=-1,expand=True)
sistolica_diastolica
```

```python
# atribuindo nomes às colunas com os valores separados
sistolica_diastolica.columns = ["Pressão Sistólica", "Pressão Diastólica"]
```

```python
# convertendo os valores para o tipo inteiro
sistolica_diastolica = sistolica_diastolica.astype(int)
```

```python
sistolica_diastolica["Pressão Média"] = sistolica_diastolica["Pressão Diastólica"] + (sistolica_diastolica["Pressão Sistólica"] - sistolica_diastolica["Pressão Diastólica"]) / 3
sistolica_diastolica
# fórmula de cálculo da Pressão média
```

```python
# adicionando ao dataframe a coluna de gênero do dataframe original
sistolica_diastolica["Gênero"]=df_saude_sono["Gênero"]
sistolica_diastolica
```

```python
# Agrupando os dados por gênero e calculando a média da pressão média para cada grupo
sistolica_diastolica.groupby("Gênero")["Pressão Média"].mean()
```

Considerando a média, a pressão sanguínea média das mulheres é ligeiramente maior que a dos homens
