# Valores faltosos

15. Para os valores faltosos coloque:
- Se string: Falta informação
- Se número: mediana
16. Plote a porcentagem de dados faltosos para cada coluna.

```python
(df_concursos.isna().sum() / len(df_concursos) * 100).plot(kind="bar", figsize=(10,5), title="Porcentagem de dados faltantes por coluna")
#soma a quantidade de dados faltantes de cada coluna e calcula a porcentagem relativa, e cria gráfico de barras para cada coluna
```

```python
col_string = df_concursos.select_dtypes(include='object').columns
#cria uma variável chamada col_string, que vai armazenar as colunas do tipo texto
```

```python
df_concursos[col_string] = df_concursos[col_string].fillna('Falta informação')
#substitui nas colunas armazenadas em col_string todos os dados faltantes por "Falta informação"
```

```python
col_num = df_concursos.select_dtypes(include='number').columns
#cria uma variável chamada col_num, que vai armazenar as colunas do tipo numéricos
```

```python
df_concursos[col_num]=df_concursos[col_num].fillna(df_concursos[col_num].median())
##substitui nas colunas armazenadas em col_num todos os dados faltantes pela mediana dos dados
```
